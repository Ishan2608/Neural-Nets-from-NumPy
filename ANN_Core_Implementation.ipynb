# 3. The Building Blocks of the ANN

## 3.1. Weights and Bias Matrix

- In an ANN, every input value is mapped to every node of the next layer. Thus, for $N$ input values and $M$ nodes in the next layer, we require a weight matrix $W$ of size $N x M$. Note that each $N$ corresponds to each feature in the dataset.

- The $ij$-th element of the weight matrix, denoted as $W_{ij}$, represents the weight connecting the $i$-th input to the $j$-th node of the next layer.

- From above point, understand that first row $i=0$ of the matrix $W$, represents all weights associated with the first input neuron to every neuron of the next layer. Similarly, the second row represents all weights associated to the second neuron.

- We also conclude that each column $j$ of the matrix $W$, represents each `feature` of the dataset.

- In addition to weights, each node in a layer has a bias term. The bias allows the activation function to be shifted, helping the model to fit the data better.

- For a layer with $M$ nodes, we require a bias vector $b$ of size $1 x M$ (or $M x 1$, depending on convention). Each element $b_j$ corresponds to the bias of the $j$-th node.

- The bias is applied to the weighted sum of the inputs before the activation function is applied.

## 3.2. Activation Function

```python
# Relu Activation Function
def relu(x):
    return np.maximum(0, x)

# Derivative of Relu Activation Function
def relu_derivative(x):
    return np.where(x > 0, 1, 0)
```

```python
print(f"X = -10: {relu(-10)}")
print(f"X = 10: {relu(10)}")
print(f"X = 0: {relu(0)}")
```

```python
# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of Sigmoid Activation Function
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)
```

```python
print(f"X = -10: {sigmoid(-10)}")
print(f"X = 0: {sigmoid(0)}")
print(f"X = 10: {sigmoid(10)}")
```

## 3.3. Loss Functions

### For Regression

```python
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def mse_derivative(y_true, y_pred):
    return 2 * (y_pred - y_true) / len(y_true)
```

### For Classification

```python
def binary_cross_entropy_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

def binary_cross_entropy_derivative(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    # Derivative of BCE w.r.t. y_pred
    return (-(y_true / y_pred) + (1 - y_true) / (1 - y_pred)) / len(y_true)
```

## 3.4. Forward Propagation

$Z = np.dot(X, W) + b$

### What Happens?

$n = number of samples$ <br>
$N = number of features for each sample$
$M = number of neurons in next layers
- X has the following shape = (n, N)

- W has the shape = (N, M)

- Dot product takes each sample, which is a row vector of size (1, N), and multiplies it with the weight matrix W (N, M). This results in a vector of size (1, M) for each sample, representing the weighted sum of inputs for each neuron in the next layer.

- Adding the bias vector b (1, num_neurons) to this weighted sum gives the input to the activation function for each neuron.

### Understanding How it Happens?

- Let's break down how the dot product `np.dot(X, W)` represents the weighted sum and how you can visualize it in the context of a neural network layer.

- Remember the formula for the weighted sum for a *single* neuron: $z = \sum_i (w_i * x_i) + b$. The dot product extends this concept to an entire layer of neurons for multiple samples simultaneously.

- Imagine your input matrix `X` (shape `n x N`) where each row is a single sample (`1 x N`), and your weight matrix `W` (shape `N x M`) where each column represents the weights connecting the inputs to a specific neuron in the next layer.

- When you perform the matrix multiplication `np.dot(X, W)`, the value in the $jk$-th position of the resulting matrix $Z$ is calculated by taking the dot product of the $j$-th row of $X$ (which is the $j$-th sample) and the $m$-th column of $W$ (which are the weights for the $m$-th neuron).

Here's how to visualize it:

- For the first sample (the first row of X) and the first neuron (the first column of W):
The dot product is: $X_{1,1}*W_{1,1} + X_{1,2}*W_{2,1} + ... + X_{1,N}*W_{N,1}$. This is exactly the weighted sum of the first sample's features with the weights of the first neuron.

- This process is repeated for every sample (row in X) and every neuron (column in W).

- So, the resulting matrix $Z$ (shape `n x M`) will have:
    - Each row representing a sample.
    - Each column representing a neuron in the next layer.
- Each element $Z_{j,m}$ being the weighted sum of the $j$-th sample's inputs using the weights of the $m$-th neuron.

- After this dot product, you add the bias vector `b` (shape `1 x M`) to each row of `Z`. This broadcast operation effectively adds the bias of each neuron to its corresponding weighted sum for every sample.

- So, `Z = np.dot(X, W) + b` efficiently calculates the weighted sum plus bias for all neurons in the layer, for all samples in your batch, in one operation.

### Code Implementation

```python
def forward(X, weights, biases, activations):
    """
    Performs the forward pass through the neural network.
    Args:
        X (np.ndarray): Input data.
        weights (list): List of weight matrices for each layer.
        biases (list): List of bias vectors for each layer.
        activations (list): List of activation functions for each layer.
    Returns:
        tuple: A tuple containing:
            - A_out (np.ndarray): The output of the last layer.
            - caches (list): A list of dictionaries, where each dictionary contains 'Z' (weighted sum) and 'A' (activation output) for each layer.
    """
    A = X
    caches = []
    num_layers = len(weights)

    for i in range(num_layers):
        W = weights[i]
        b = biases[i]
        activation_func = activations[i]

        Z = np.dot(A, W) + b
        A = activation_func(Z)

        caches.append({"Z": Z, "A": A})

    A_out = A # The output of the last layer

    return A_out, caches
```

- **Initialization:**
  - `A = X`: The input to the first layer is the original input data `X`.
  - `caches = []`: An empty list is initialized to store the intermediate results (weighted sums `Z` and activations `A`) of each layer during the forward pass. These are needed for backpropagation.
  - `num_layers = len(weights)`: The number of layers in the network is determined by the number of weight matrices provided.

- **Looping Through Layers:**
  - `for i in range(num_layers):`: The code iterates through each layer of the network. `i` represents the current layer index.

- **Getting Layer Parameters and Activation:**
  - `W = weights[i]`: The weight matrix for the current layer `i`.
  - `b = biases[i]`: The bias vector for the current layer `i`.
  - `activation_func = activations[i]`: The activation function for the current layer `i`.

- **Calculating Weighted Sum (Z):**
  - `Z = np.dot(A, W) + b`: This is the core calculation for each layer. It performs the matrix multiplication of the input from the previous layer (`A`) with the current layer's weight matrix (`W`) and then adds the bias vector (`b`).
    - **What is happening?** We are computing the linear combination of the inputs from the previous layer, scaled by the weights of the current layer, and shifted by the bias.
    - **Why is it happening this way?** This operation represents the transformation of the input signal within a neuron before the non-linear activation is applied. The dot product efficiently calculates the sum of (input * weight) for all inputs and neurons in the layer, and the bias shifts the activation boundary.

  - Let's denote the input from the previous layer as $A^{(i-1)}$ (with $A^{(0)} = X$), the weights of the current layer as $W^{(i)}$, and the biases as $b^{(i)}$. The weighted sum for layer $i$ is calculated as:
    - $$Z^{(i)} = A^{(i-1)} W^{(i)} + b^{(i)}$$
    - If $A^{(i-1)}$ has shape $(n\_{samples}, n\_{features}^{(i-1)})$ and $W^{(i)}$ has shape $(n\_{features}^{(i-1)}, n\_{neurons}^{(i)})$, then the dot product $A^{(i-1)} W^{(i)}$ will have shape $(n\_{samples}, n\_{neurons}^{(i)})$.
    - The bias $b^{(i)}$ is typically a row vector of shape $(1, n\_{neurons}^{(i)})$ and is broadcast across all samples.

- **Applying Activation Function (A):**
  - `A = activation_func(Z)`: The calculated weighted sum `Z` is passed through the activation function defined for the current layer.
    - **What is happening?** A non-linear transformation is applied to the weighted sum.
    - **Why is it happening this way?** Activation functions introduce non-linearity into the network, allowing it to learn complex patterns that linear models cannot. The result `A` becomes the input to the next layer.

    The activation for layer $i$ is calculated as:
    $$A^{(i)} = \text{activation\_func}(Z^{(i)})$$

- **Storing Intermediate Values:**
  - `caches.append({"Z": Z, "A": A})`: A dictionary containing the weighted sum (`Z`) and the activation output (`A`) for the current layer is appended to the `caches` list.
    - **What is happening?** We are saving the results of the linear and non-linear transformations for each layer.
    - **Why is it happening this way?** These values (specifically `Z` and the `A` from the *previous* layer, which is the input to the current layer's calculation of `Z`) are essential for calculating the gradients during the backward propagation process using the chain rule.

- **Final Output:**
  - `A_out = A`: After the loop finishes, the variable `A` holds the activation output of the last layer, which is the final output of the neural network.

- **Returning Values:**
  - `return A_out, caches`: The function returns the final output of the network (`A_out`) and the list of cached intermediate values (`caches`). The `caches` are passed to the backward propagation function.

## 3.5. Backward Propagation

Backward propagation is the process of calculating the gradients of the loss function with respect to the weights and biases of the neural network. These gradients indicate how much each parameter contributed to the error, allowing us to adjust the parameters to minimize the loss during training.

### The Chain Rule

- The core principle behind backward propagation is the **chain rule** from calculus. Since the output of the network is a function of the weights and biases through multiple layers and activation functions, we use the chain rule to calculate the gradient of the loss function with respect to each parameter.

- If we have a loss function $L$ that depends on the output $y$, which in turn depends on a weight $w$, the chain rule states:

- $\frac{\partial L}{\partial w} = \frac{\partial L}{\partial y} \times \frac{\partial y}{\partial w}$

- In a neural network, this chain extends through multiple layers.



---



### Calculating the Output Layer Gradients

- Let's consider the output layer first. The loss function, for example, Mean Squared Error (MSE) for regression or Binary Cross-Entropy (BCE) for classification, measures the difference between the network's output and the true target.

- For a regression problem with MSE loss:
$L = \frac{1}{n} \sum_{i=1}^{n} (y_{pred}^{(i)} - y_{true}^{(i)})^2$

- The gradient of the loss with respect to the output of the last layer ($A_{out}$) is:
$\frac{\partial L}{\partial A_{out}} = \frac{2}{n} (A_{out} - y_{true})$

- To get the gradient with respect to the weights ($W_{out}$) and biases ($b_{out}$) of the output layer, we apply the chain rule:

    - $\frac{\partial L}{\partial W_{out}} = \frac{\partial L}{\partial A_{out}} \times \frac{\partial A_{out}}{\partial Z_{out}} \times \frac{\partial Z_{out}}{\partial W_{out}}$
    - $\frac{\partial L}{\partial b_{out}} = \frac{\partial L}{\partial A_{out}} \times \frac{\partial A_{out}}{\partial Z_{out}} \times \frac{\partial Z_{out}}{\partial b_{out}}$

- Where $Z_{out}$ is the weighted sum before the activation function in the output layer, and $\frac{\partial A_{out}}{\partial Z_{out}}$ is the derivative of the output activation function.

- For example, if the output activation is sigmoid, $\frac{\partial A_{out}}{\partial Z_{out}} = A_{out}(1 - A_{out})$.

- The partial derivatives with respect to $W_{out}$ and $b_{out}$ are:

    - $\frac{\partial Z_{out}}{\partial W_{out}} = X_{last\_hidden\_layer}$ (Input from the previous layer)
    - $\frac{\partial Z_{out}}{\partial b_{out}} = 1$

- So, the gradients for the output layer are:

    - $\frac{\partial L}{\partial W_{out}} = \frac{\partial L}{\partial A_{out}} \times \text{derivative of output activation} \times X_{last\_hidden\_layer}^T$
    - $\frac{\partial L}{\partial b_{out}} = \frac{\partial L}{\partial A_{out}} \times \text{derivative of output activation}$



---



### Backpropagating the Error

- The error calculated at the output layer needs to be propagated backward to the hidden layers to calculate their gradients. The amount of error attributed to each neuron in the hidden layer is proportional to the weights connecting it to the next layer and the error in that layer.

- The error signal for a hidden layer ($Error_{hidden}$) is calculated by taking the dot product of the error in the next layer ($Error_{next}$) and the transpose of the weights connecting the hidden layer to the next layer ($W_{next}^T$), and then multiplying by the derivative of the hidden layer's activation function:

- $Error_{hidden} = (Error_{next} \cdot W_{next}^T) \times \text{derivative of hidden activation}$



---



### Calculating Hidden Layer Gradients

- Once the error signal for a hidden layer is calculated, we can find the gradients for its weights ($W_{hidden}$) and biases ($b_{hidden}$):

    - $\frac{\partial L}{\partial W_{hidden}} = X_{previous\_layer}^T \cdot Error_{hidden}$
    - $\frac{\partial L}{\partial b_{hidden}} = Error_{hidden}$

    - Where $X_{previous\_layer}$ is the output from the preceding layer (or the original input X for the first hidden layer).



---



### Weight and Bias Updates

- Finally, the calculated gradients are used to update the weights and biases using an optimization algorithm like gradient descent. The update rule for a parameter ($\theta$, which can be a weight or a bias) is:

- $\theta = \theta - \alpha \times \frac{\partial L}{\partial \theta}$

- Where $\alpha$ is the learning rate, a hyperparameter that controls the step size of the update.



---



### Putting it all together

Backward propagation is an iterative process. During each training epoch, after a forward pass calculates the output and the loss, backward propagation calculates the gradients for all parameters. These gradients are then used to update the parameters, and the process repeats, gradually minimizing the loss and improving the network's performance.

### Code Implementation

```python
def backward(X, y_true, y_pred, caches, weights, activations_derivative, problem_type):
    gradients = {}
    num_layers = len(weights)
    n_samples = X.shape[0]

    # Start backpropagation from the output layer
    output_layer_cache = caches[num_layers - 1]
    Z_out = output_layer_cache["Z"]
    A_out = output_layer_cache["A"] # This is y_pred

    # Calculate the initial gradient dL/dZ_out based on problem type
    if problem_type == 'regression':
        # For MSE loss and linear output
        dL_dZ_out = (A_out - y_true) * (2 / n_samples)
    elif problem_type == 'classification':
        # For Binary Cross-Entropy with Sigmoid output
        dL_dZ_out = (A_out - y_true) / n_samples
    else:
        raise ValueError("problem_type must be 'regression' or 'classification'")

    # Calculate gradients for the output layer weights and biases
    # The input to the output layer is the activation of the last hidden layer (A_last_hidden)
    # If there's only one layer (no hidden layers), the input is X
    A_prev = caches[num_layers - 2]["A"] if num_layers > 1 else X

    gradients[f"dW{num_layers}"] = np.dot(A_prev.T, dL_dZ_out)
    gradients[f"db{num_layers}"] = np.sum(dL_dZ_out, axis=0, keepdims=True)

    # Backpropagate through hidden layers
    # Iterate from the second to last layer down to the first layer
    for i in range(num_layers - 2, -1, -1):
        current_layer_cache = caches[i]
        Z_current = current_layer_cache["Z"]
        A_current = current_layer_cache["A"] # Output of current layer

        # Get the weight matrix of the next layer (which we just backpropagated through)
        W_next = weights[i + 1]

        # Calculate the gradient of the loss with respect to the current layer's activation (dA_current)
        # dL/dA_current = (dL/dZ_next * W_next.T)
        dL_dA_current = np.dot(dL_dZ_out, W_next.T)

        # Get the derivative of the current layer's activation function
        # Use the appropriate derivative from the provided list
        current_activation_derivative = activations_derivative[i]

        # Calculate the gradient of the loss with respect to the current layer's weighted sum (dZ_current)
        # dL/dZ_current = dL/dA_current * derivative of current activation
        dL_dZ_current = dL_dA_current * current_activation_derivative(Z_current)

        # Calculate gradients for the current layer's weights and biases
        # The input to the current layer is the activation of the previous layer (A_prev_layer)
        # If it's the first hidden layer (i = 0), the input is X
        A_prev_layer = caches[i - 1]["A"] if i > 0 else X

        gradients[f"dW{i + 1}"] = np.dot(A_prev_layer.T, dL_dZ_current)
        gradients[f"db{i + 1}"] = np.sum(dL_dZ_current, axis=0, keepdims=True)

        # Set dL_dZ_out for the next iteration (which will be the previous layer)
        dL_dZ_out = dL_dZ_current

    return gradients
```

The `backward` function performs the backward pass of a multi-layer neural network. It calculates the gradients of the `MSE` loss function or the `BCE` loss function with respect to the weights and biases of each layer, using the chain rule and the intermediate values stored during the forward pass (`caches`). These gradients are then used to update the network's parameters during training.

Here's a breakdown of the function:

- **Initialization:**
  - `gradients = {}`: An empty dictionary is initialized to store the calculated gradients for the weights and biases of each layer.
  - `num_layers = len(weights)`: The total number of layers in the network (including the output layer) is determined.
  - `n_samples = X.shape[0]`: The number of samples in the input data, used for averaging gradients over the batch.

- **Starting Backpropagation from the Output Layer:**
  - `output_layer_cache = caches[num_layers - 1]`: Retrieves the cached values (Z and A) for the output layer.
  - `Z_out = output_layer_cache["Z"]`: The weighted sum before the activation in the output layer.
  - `A_out = output_layer_cache["A"]`: The output of the output layer after activation (which is also `y_pred` in this case).

  - **Calculate the gradient of the loss with respect to the output layer's weighted sum (dL_dZ_out):**
    - `dL_dZ_out = (A_out - y_true) * (2 / n_samples)`: This calculates the gradient of the Mean Squared Error (MSE) loss with respect to the weighted sum ($Z$) of the output layer. For MSE with a linear output layer, the derivative of the loss with respect to the output ($A_{out}$) is $\frac{2}{n}(A_{out} - y_{true})$, and since the derivative of the linear activation is 1, the gradient with respect to $Z_{out}$ is the same. We divide by `n_samples` to get the mean gradient over the batch.
    - **What is happening?** We are determining how much the loss changes with respect to the values just before the output activation.
    - **Why is it happening this way?** This is the starting point of the backpropagation. We need to know how the error at the output relates to the values produced by the last layer's linear transformation.

- **Calculate Gradients for the Output Layer Weights and Biases:**
  - `A_prev = caches[num_layers - 2]["A"] if num_layers > 1 else X`: Gets the activation output of the layer *before* the output layer. If there's only one layer (no hidden layers), the input to the output layer is the original input `X`.
  - `gradients[f"dW{num_layers}"] = np.dot(A_prev.T, dL_dZ_out)`: Calculates the gradient of the loss with respect to the output layer's weights ($W_{out}$). This is done by multiplying the transpose of the input to the output layer ($A_{prev}$) with the gradient of the loss with respect to $Z_{out}$ ($dL\_dZ_{out}$).
    - $\frac{\partial L}{\partial W^{(num\_layers)}} = (A^{(num\_layers-1)})^T \cdot \frac{\partial L}{\partial Z^{(num\_layers)}}$
  - `gradients[f"db{num_layers}"] = np.sum(dL_dZ_out, axis=0, keepdims=True)`: Calculates the gradient of the loss with respect to the output layer's biases ($b_{out}$). This is the sum of the gradient of the loss with respect to $Z_{out}$ across all samples.
    - $\frac{\partial L}{\partial b^{(num\_layers)}} = \sum_{samples} \frac{\partial L}{\partial Z^{(num\_layers)}}$

- **Backpropagate Through Hidden Layers:**
  - `for i in range(num_layers - 2, -1, -1):`: This loop iterates backward through the hidden layers, starting from the last hidden layer (`num_layers - 2`) up to the first hidden layer (`0`).
  - `current_layer_cache = caches[i]`: Retrieves the cached values (Z and A) for the current hidden layer.
  - `Z_current = current_layer_cache["Z"]`: The weighted sum before the activation in the current hidden layer.
  - `A_current = current_layer_cache["A"]`: The output of the current hidden layer after activation.

  - `W_next = weights[i + 1]`: Gets the weight matrix of the layer *after* the current hidden layer (i.e., the layer we just backpropagated through).

  - **Calculate the gradient of the loss with respect to the current layer's activation (dL_dA_current):**
    - `dL_dA_current = np.dot(dL_dZ_out, W_next.T)`: This calculates how much the loss changes with respect to the output of the current hidden layer. It's done by multiplying the gradient from the *next* layer ($dL\_dZ_{out}$) with the transpose of the weights connecting the current layer to the next layer ($W_{next}^T$).
    - $\frac{\partial L}{\partial A^{(i)}} = \frac{\partial L}{\partial Z^{(i+1)}} \cdot (W^{(i+1)})^T$

  - **Get the derivative of the current layer's activation function:**
    - `current_activation_derivative = activations_derivative[i]`: Retrieves the derivative function for the activation function used in the current hidden layer.

  - **Calculate the gradient of the loss with respect to the current layer's weighted sum (dL_dZ_current):**
    - `dL_dZ_current = dL_dA_current * current_activation_derivative(Z_current)`: This applies the chain rule to find how the loss changes with respect to the weighted sum of the current hidden layer. It's the product of the gradient with respect to the activation ($dL\_dA_{current}$) and the derivative of the activation function evaluated at the weighted sum ($Z_{current}$).
    - $\frac{\partial L}{\partial Z^{(i)}} = \frac{\partial L}{\partial A^{(i)}} \odot \text{activation\_derivative}(Z^{(i)})$ (where $\odot$ is the element-wise product)

  - **Calculate Gradients for the current layer's Weights and Biases:**
    - `A_prev_layer = caches[i - 1]["A"] if i > 0 else X`: Gets the activation output of the layer *before* the current hidden layer. If it's the first hidden layer (i=0), the input is the original input `X`.
    - `gradients[f"dW{i + 1}"] = np.dot(A_prev_layer.T, dL_dZ_current)`: Calculates the gradient of the loss with respect to the current hidden layer's weights ($W_{current}$).
    - $\frac{\partial L}{\partial W^{(i+1)}} = (A^{(i)})^T \cdot \frac{\partial L}{\partial Z^{(i+1)}}$ (Note the index mapping: $A^{(i)}$ is the input to layer $i+1$, and $W^{(i+1)}$ is the weight matrix for layer $i+1$)
    - `gradients[f"db{i + 1}"] = np.sum(dL_dZ_current, axis=0, keepdims=True)`: Calculates the gradient of the loss with respect to the current hidden layer's biases ($b_{current}$).
    - $\frac{\partial L}{\partial b^{(i+1)}} = \sum_{samples} \frac{\partial L}{\partial Z^{(i+1)}}$

  - `dL_dZ_out = dL_dZ_current`: Updates `dL_dZ_out` to be the gradient of the loss with respect to the weighted sum of the *current* layer. In the next iteration of the loop, this will be used to calculate the gradient for the *previous* layer, effectively propagating the error backward.

- **Return Gradients:**
  - `return gradients`: The function returns the dictionary containing all the calculated gradients for the weights and biases.

### Explanation of `A_out - y_true` in Classification Backpropagation

- You asked why we use `dL_dZ_out = (A_out - y_true) / n_samples` in the `backward_classification` function, given that the target classes are 0 or 1. This is a common point of confusion, but it is mathematically correct due to the relationship between the Binary Cross-Entropy (BCE) loss and the Sigmoid activation function.

Here's the simplified explanation:

- The gradient of the Binary Cross-Entropy loss ($L$) with respect to the output of the sigmoid activation ($A_{out}$) is:
$$ \frac{\partial L}{\partial A_{out}} = -\left(\frac{y_{true}}{A_{out}} - \frac{1 - y_{true}}{1 - A_{out}}\right) $$

- The derivative of the sigmoid activation function ($A_{out}$) with respect to its input ($Z_{out}$) is:
$$ \frac{\partial A_{out}}{\partial Z_{out}} = A_{out}(1 - A_{out}) $$

- Using the chain rule, the gradient of the loss with respect to the weighted sum ($Z_{out}$) is:
$$ \frac{\partial L}{\partial Z_{out}} = \frac{\partial L}{\partial A_{out}} \times \frac{\partial A_{out}}{\partial Z_{out}} $$
- When you substitute the derivatives and simplify the expression, you find that:
$$ \frac{\partial L}{\partial Z_{out}} = A_{out} - y_{true} $$

- So, while it looks like a simple subtraction of a true class from a predicted probability, it's the result of applying the chain rule through the BCE loss and the sigmoid activation. This simplification makes the backward pass calculation very efficient for this common combination of loss and activation.

- The division by `n_samples` is included to calculate the mean gradient over the batch, which is standard practice during training with gradient descent.

## 3.6. Initialize and Update Parameters

```python
def initialize_parameters(layers):
    """
    He initialization for hidden layers (good with ReLU).
    layers: list of layer sizes, e.g. [input_dim, ... , output_dim]
    """
    weights = []
    biases = []
    num_of_layers = len(layers)

    for i in range(1, num_of_layers):
        fan_in = layers[i-1]
        # He init (suitable when hidden layers use ReLU)
        std = np.sqrt(2.0 / fan_in)
        w = np.random.randn(fan_in, layers[i]) * std
        b = np.zeros((1, layers[i]))

        weights.append(w)
        biases.append(b)
    return (weights, biases)
```

```python
def update_parameters(weights, biases, gradients, learning_rate):
    """
    Updates weights and biases using gradient descent.
    Args:
        weights (list): List of weight matrices.
        biases (list): List of bias vectors.
        gradients (dict): Dictionary containing gradients for weights and biases.
        learning_rate (float): The learning rate for gradient descent.
    Returns:
        tuple: A tuple containing the updated weights and biases.
    """
    num_layers = len(weights)

    for i in range(num_layers):
        weights[i] -= learning_rate * gradients[f"dW{i + 1}"]
        biases[i] -= learning_rate * gradients[f"db{i + 1}"]

    return weights, biases
```

## 3.7. Training Neural Networks

```python
def train_NN(X, y, layers, activations, activations_derivs, alpha, epochs=100, problem_type='regression', print_loss=True):

    Ws, Bs = initialize_parameters(layers)
    losses = []

    # Reshape y_true for consistent dimensions - (n, 1)
    y = y.reshape(-1, 1)

    for epoch in range(1, epochs + 1):
        # Forward Pass

        y_pred, caches = forward(X, Ws, Bs, activations)
        # Calculate Loss
        if problem_type == 'regression':
            loss = mse_loss(y, y_pred)
        elif problem_type == 'classification':
            loss = binary_cross_entropy_loss(y, y_pred)

        # Backward pass
        gradients = backward(X, y, y_pred, caches, Ws, activations_derivs, problem_type)

        # Update parameters
        weights, biases = update_parameters(Ws, Bs, gradients, alpha)

        # Print loss
        if print_loss and epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss:.4f}")
            losses.append(loss)

    return weights, biases, losses
```

## 3.8. Prediction Function

```python
def predict(X, weights, biases, activations):
    # The prediction process is essentially a forward pass
    y_pred, _ = forward(X, weights, biases, activations)
    return y_pred
```
